```python
import pandas as pd
import numpy as np
import plotly.express as px
from datetime import datetime
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('c-rate_regdata.csv')
df
```

```python
df.columns
df['state'].unique()
```

```python
df = df[(df['state.1']==3) & (df['battcurrent']<=0)]
df['calc_current'] = df['battcurrent']/144
group_df = df.groupby(['vin','state','cyclenum']).agg({'calc_current':'mean', 'odometer':'min', 'intime':'min'}).reset_index()
```

```python
group_df
```

```python
vin_group = group_df.groupby(['vin'])
type(group_df['intime'][0])
```

## filter for derived battcurrent < -0.01
### remove all positive values from battcurrent before taking avg only



```python


def plot(index):

    filtered_df = [vin_group.get_group(x) for x in vin_group.groups][index].copy()
    print(filtered_df)
    vin = filtered_df.vin.unique()[0]
    filtered_df = filtered_df[filtered_df['calc_current'] <= -0.01]
    filtered_df = filtered_df.sort_values('intime').reset_index(drop = True)
    filtered_df['intime'] = pd.to_datetime(filtered_df['intime'], infer_datetime_format=True)
    filtered_df['date'] = filtered_df['intime'].apply(lambda x : str(x.day)+'-'+str(x.month)+'-'+str(x.year))
    fig = px.line(filtered_df, x="date", y="calc_current", title=vin+" battery current TREND")
    fig.show()
    
list(map(plot, range(vin_group.ngroups)))    
```
